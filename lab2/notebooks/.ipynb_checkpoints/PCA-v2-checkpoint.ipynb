```python
from sklearn.decomposition import PCA
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

We are using the training set 1 because it only contains normal behavior

```python
file_name = "trn_set_1"
DATA_PATH = r"C:../data/{}.csv".format(file_name)

data_trn_1 = pd.read_csv(DATA_PATH, header=0, parse_dates=['DATETIME'], index_col='DATETIME', dayfirst=True)
data_trn_1.index = pd.DatetimeIndex(data_trn_1.index.values, freq = data_trn_1.index.inferred_freq)

discrete_variables = [f'S_PU{num+1}' for num in range(11)] + ['S_V2']  

# drop discrete variables
data_trn_1 = data_trn_1.drop(columns=discrete_variables)
data_trn_1 = data_trn_1.drop(columns='ATT_FLAG') # labels are useless for first dataset
```

First of all... NORMALIZATION!

```python
df_mean_trn = data_trn_1.mean()
df_std_trn = data_trn_1.std()
normalized_df=(data_trn_1-df_mean_trn)/df_std_trn
# display(normalized_df.head())

# remove columns with NaN (they had only the same value --> not possible to normalize them)
print(normalized_df.columns[normalized_df.isna().any()].tolist())
# spoiler: we will need to remove these columns also from the testset when we will do anomaly detection


normalized_df.dropna(axis='columns', inplace=True)
# normalized_df.head()
```

```python
n_of_components = normalized_df.shape[1]
pca_full = PCA(n_of_components).fit(normalized_df)

plt.figure()
plt.plot(np.cumsum(pca_full.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)')
plt.title('Explained variance (on normalized dataset)')
plt.show()
```

Let's create a function that retrieves the number of principal components that we need if we want to capture a certain percentage of variance in the data

```python
def get_num_of_principal_components(fitted_pca, desired_variance_perc):
    cumulative_variance = np.cumsum(fitted_pca.explained_variance_ratio_)
    return [i for i in range(len(cumulative_variance)) if cumulative_variance[i] > desired_variance_perc/100][0]
```

Let's use this number of components to project the data into a lower dimensional feature space. In order to capture the normal behaviour (and therefore being able to detect anomalies) we need to capture enough variance of the data, but not too much. Otherwise we would also learn about the anomalies. This is actually the power of PCA!

```python
var_perc_to_capture = 99  # variance that we want to capture to learn the normal behavior

n = get_num_of_principal_components(pca_full, var_perc_to_capture)
print(f"{n} components out of {normalized_df.shape[1]} will be used to project into the normal space")

pca_normal = PCA(n_components=n).fit(normalized_df)
```

# Training set 2

We will use it to tune the threshold on the residuals, obtained with the PCA projection learned on the normal dataset (training set 1)

```python
file_name = "trn_set_2"
DATA_PATH = r"C:../data/{}.csv".format(file_name)

data_trn_2 = pd.read_csv(DATA_PATH, header=0, parse_dates=['DATETIME'], index_col='DATETIME', dayfirst=True)
data_trn_2.index = pd.DatetimeIndex(data_trn_2.index.values, freq = data_trn_2.index.inferred_freq)

labels = data_trn_2['ATT_FLAG']

# normalize using mean and standard deviation from the training set (https://www.researchgate.net/post/If_I_used_data_normalization_x-meanx_stdx_for_training_data_would_I_use_train_Mean_and_Standard_Deviation_to_normalize_test_data)
data_trn_2 = (data_trn_2-df_mean_trn)/df_std_trn

# remove also features which could not be normalized when training PCA on the 'normal' dataset
data_trn_2 = data_trn_2[list(normalized_df.columns)] 
data_trn_2.head()
```

```python
# we are going to use the pca that we obtained in the training set to see if we can detect anomalies in the test set
data_trn_2_pca = pd.DataFrame(pca_normal.transform(data_trn_2), data_trn_2.index)
data_trn_2_pca.head()
```

```python
# reconstruct dataframe
data_trn_2_reconstr = pd.DataFrame(pca_normal.inverse_transform(data_trn_2_pca), data_trn_2_pca.index)
data_trn_2_reconstr.columns = data_trn_2.columns
data_trn_2_reconstr.head()
```

```python
# Now, we compute the residuals: original normalized data - reconstructed data
residual_trn_2 = data_trn_2.subtract(data_trn_2_reconstr)
squared_res_trn_2 = pd.DataFrame(np.power(residual_trn_2, 2), residual_trn_2.index)
squared_sum_res_trn_2 = squared_res_trn_2.sum(axis=1)
```

```python
# Let's visualize it!

att_true = int(np.max(squared_sum_res_trn_2) + 10)
att_false = int(np.min(squared_sum_res_trn_2) - 10)
labels_remap = labels.map({-999:att_false, 1:att_true})

plt.figure(figsize=(17,3))
plt.plot(squared_sum_res_trn_2, label='Sum of residuals')
plt.plot(labels_remap, label="True labels", alpha=0.5)
plt.legend()
plt.yscale("log")
plt.show()
```

It is clear that the PCA learned on the normal behaviour, is not able to reconstruct anomalies, and we can therefore find them by plotting the sum of the squared residuals. Results look really promising! Let's now do some gridsearch to find a treshold on the residuals.

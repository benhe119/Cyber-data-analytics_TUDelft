```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's define a nice way of plotting the correlation matrix using seaborn

```python
def plot_corr_sns(df, file_name):
    sns.set()
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(10, 10))
    ax = sns.heatmap(
            corr, 
            cmap=sns.diverging_palette(20, 220, n=200),
            square=True,
    )
    ax.set_xticklabels(
        ax.get_xticklabels(),
        rotation=90,
        fontsize=10
    )
    ax.set_yticklabels(
        ax.get_yticklabels(),
        fontsize=10
    )
    plt.title(file_name)
    plt.savefig(r"../images/correlation_matrix_{}.pdf".format(file_name))
    plt.show()
```

Let's start by analysing correlations between the signals. We found some whitespaces in the column names (training set 2) and therefore we fixed them.

```python
for file_name in ['trn_set_1', 'trn_set_2', 'test_set']:
    DATA_PATH = r"C:../data/{}.csv".format(file_name)
    data = pd.read_csv(DATA_PATH, header=0, parse_dates=['DATETIME'], index_col='DATETIME', dayfirst=True)

    
    discrete_variables = [f'S_PU{num+1}' for num in range(11)] + ['S_V2']
    
    data_no_discr = data.drop(columns=discrete_variables)
    plot_corr_sns(data_no_discr, file_name)
```

### Prediction using "persistence"

```python
from sklearn.metrics import mean_squared_error
```

```python
# We choose 'L_T1' as the sensor to validate the 'persistence' prediction
# data.loc['2016-10-01':'2016-11-01'][cols_plot]
timeseries = data['L_T1']
print(timeseries.shape)

timeseries.plot(figsize=(40, 7))
plt.show() # just to have an idea of the whole shape of the sensor
```

```python
predictions = list()
delay = 1 #timesteps to delay for the prediction
for idx, x in enumerate(timeseries):
    if idx < delay:
        continue # We cannot do persistence on the first step
    yhat = timeseries[idx-delay] # we predict value at time t using value at time t-<delay>
    predictions.append(yhat)
test_score = mean_squared_error(timeseries[delay:], predictions)
print('Test MSE: %.3f' % test_score)
```

Let's see how the prediction looks like. For visualization purposes, we only show two months of predictions (The MSE is computed on the whole timeseries though)

```python
subset = 24 * 8  # this is just done for the visualization, to better see how the delay is used for the prediction
plt.figure(num=None, figsize=(15, 7))
plt.plot(timeseries.values[delay:delay+subset], color='blue', label="Observed")
plt.plot(predictions[:subset], color='red', label="Predicted")
plt.legend()
plt.show()
```

We can see how each prediction (red) is just the previous observed value (in blue).
**Prediction is easy for this signal.**
